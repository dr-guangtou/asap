```python
from umachine_pyio.load_mock import load_mock_from_binaries, value_added_mock

galprops = list((
    'sm', 'sfr', 'obs_sm', 'obs_sfr', 'icl', 'halo_id', 'upid',
    'x', 'y', 'z', 'vx', 'vy', 'vz', 'rvir', 'mvir', 'mpeak'))
Lbox = 250.
mock = value_added_mock(load_mock_from_binaries(range(144), galprops=galprops), Lbox)
mock['m100'] = mock['icl'] + mock['sm']

mask = mock['m100'] > 10**11.25
sample = mock[mask]
```

```python
from ellipse_selection_functions import ellipse_selector, split_sample_along_axis_angle

xpts, ypts = np.log10(sample['sm']), np.log10(sample['m100'])
xc, yc = 11.4, 11.7
major_axis_angle, b_by_a = 45, 20
nkeep = 500
idx_selection = ellipse_selector(xpts, ypts, xc, yc, major_axis_angle, b_by_a, nkeep)

subsample = sample[idx_selection]

xpts2, ypts2 = np.log10(subsample['sm']), np.log10(subsample['m100'])
splitting_mask = split_sample_along_axis_angle(xpts2, ypts2, xc, yc, major_axis_angle)
subsample1 = subsample[splitting_mask]
subsample2 = subsample[~splitting_mask]


fig, ax = plt.subplots(1, 1)

__=ax.scatter(np.log10(sample['sm']), np.log10(sample['m100']), s=1)
__=ax.scatter(np.log10(subsample['sm']), np.log10(subsample['m100']), s=1)
__=ax.scatter(np.log10(subsample1['sm']), np.log10(subsample1['m100']), s=1)
__=ax.scatter(np.log10(subsample2['sm']), np.log10(subsample2['m100']), s=1)

xlim = ax.set_xlim(10.5, 12.25)
ylim = ax.set_ylim(11, 12.75)
xlabel = ax.set_xlabel(r'$M_{100}$', size=15)
ylabel = ax.set_ylabel(r'$M_{10}$', size=15)

fig.savefig('selection1.png', bbox_extra_artists=[xlabel, ylabel], bbox_inches='tight')
```
